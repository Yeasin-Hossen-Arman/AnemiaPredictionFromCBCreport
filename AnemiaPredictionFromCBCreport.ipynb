1. Fever
#### •	Anemia: Can occur in severe cases caused by infections or inflammation.
#### •	Dengue & Malaria: Both are classic causes of high fever due to viral (dengue) or parasitic (malaria) infections.
#### •	Pneumonia: Often accompanied by fever as the body fights a bacterial or viral infection.
#### •	Asthma: Fever isn’t a primary symptom but may occur if there’s an underlying infection (e.g., viral-induced asthma).
________________________________________
2. Fatigue/Weakness
#### •	Anemia: A hallmark symptom due to reduced oxygen delivery to tissues.
#### •	Dengue & Malaria: Both diseases cause significant fatigue due to systemic infection and immune response.
#### •	Pneumonia: Fatigue results from the body’s effort to fight infection and reduced oxygen levels.
#### •	Asthma: Severe or poorly controlled asthma can cause fatigue due to difficulty breathing and decreased oxygen intake.
________________________________________
3. Difficulty Breathing (Dyspnea)
#### •	Anemia: In severe cases, due to lack of oxygen-carrying red blood cells.
#### •	Pneumonia: Often causes shortness of breath due to lung infection and fluid accumulation.
#### •	Asthma: A defining feature due to airway inflammation and narrowing.
#### •	Dengue & Malaria: Rarely cause primary respiratory symptoms but may lead to breathing issues in severe cases (e.g., dengue shock syndrome or severe malaria with pulmonary complications).
________________________________________
4. Headache
#### •	Anemia: Common, especially in moderate to severe cases, due to reduced oxygen to the brain.
#### •	Dengue & Malaria: Both frequently present with headaches as part of systemic symptoms.
#### •	Pneumonia: Sometimes associated with headaches, especially with fever and dehydration.
#### •	Asthma: Headache can occur due to low oxygen levels or tension from coughing.
________________________________________
5. Pale Skin
#### •	Anemia: A hallmark symptom due to reduced hemoglobin.
#### •	Dengue & Malaria: Can cause paleness in severe cases due to blood loss or hemolysis.
#### •	Pneumonia & Asthma: May cause pale or bluish skin (cyanosis) in severe cases when oxygen levels drop.
________________________________________
6. Sweating
#### •	Anemia: Can occur with exertion or as a symptom of underlying infections causing anemia.
#### •	Dengue & Malaria: Common due to high fevers and immune responses.
#### •	Pneumonia: Night sweats are a frequent symptom, particularly in bacterial infections.
#### •	Asthma: Sweating can occur during severe asthma attacks.
________________________________________
7. Rapid Heart Rate (Tachycardia)
#### •	Anemia: A compensatory response to deliver more oxygen to tissues.
#### •	Dengue & Malaria: Fever and dehydration can lead to increased heart rate.
#### •	Pneumonia: Common due to fever, hypoxia, or infection.
#### •	Asthma: May occur during an attack or due to low oxygen levels.
________________________________________
8. Chest Pain
#### •	Anemia: Rare but may occur in severe cases due to increased cardiac strain.
#### •	Pneumonia: Often associated with pleuritic chest pain (pain on breathing in).
#### •	Asthma: Chest tightness is a common symptom.
#### •	Dengue & Malaria: Can cause chest discomfort, especially if complications involve the heart or lungs.
________________________________________
Key Note:
While these conditions share symptoms, they differ in specific causes and underlying mechanisms. For instance:
#### •	Dengue and malaria are infections with systemic manifestations.
#### •	Anemia is due to low red blood cell/hemoglobin levels.
#### •	Pneumonia is a lung infection.
#### •	Asthma is a chronic inflammatory condition of the airways.


### TEST:
1. Anemia
Tests:
•	Complete Blood Count (CBC): Measures hemoglobin levels, red blood cell count, hematocrit, and mean corpuscular volume (MCV).
•	Peripheral Blood Smear: Examines the shape and size of red blood cells to diagnose specific types of anemia.
•	Iron Studies: Serum iron, ferritin, and total iron-binding capacity (TIBC) to assess iron deficiency.
•	Vitamin Tests: Vitamin B12 and folate levels for megaloblastic anemia.
•	Reticulocyte Count: Measures young red blood cells to evaluate bone marrow function.
________________________________________
2. Dengue
Tests:
•	NS1 Antigen Test: Detects the presence of the dengue virus within the first 5 days of symptoms.
•	IgM/IgG Antibody Test: Detects antibodies against the dengue virus (IgM for recent infection, IgG for past infection).
•	RT-PCR (Reverse Transcription Polymerase Chain Reaction): Identifies viral RNA for early and accurate detection.
•	Complete Blood Count (CBC): Monitors platelet count (often low in dengue) and hematocrit levels (elevated in severe cases).
________________________________________
3. Malaria
Tests:
•	Microscopic Examination: Thick and thin blood smears stained with Giemsa to identify Plasmodium parasites.
•	Rapid Diagnostic Test (RDT): Detects specific antigens from Plasmodium species.
•	Polymerase Chain Reaction (PCR): Highly sensitive for detecting low levels of malaria parasites.
•	Serology Tests: Detect antibodies to Plasmodium (less commonly used).
________________________________________
4. Pneumonia
Tests:
•	Chest X-Ray: Visualizes lung inflammation, consolidation, or fluid buildup.
•	Sputum Culture: Identifies the causative bacterial or fungal agent.
•	Blood Culture: Detects bacteria in the bloodstream.
•	Pulse Oximetry: Measures oxygen levels in the blood to assess respiratory distress.
•	C-Reactive Protein (CRP) and Procalcitonin Tests: Indicate inflammation and bacterial infection.
•	Bronchoscopy: For severe or unclear cases, to directly examine the lungs and collect samples.
________________________________________
5. Asthma
Tests:
•	Spirometry: Measures lung function, including forced expiratory volume (FEV1) and peak expiratory flow (PEF).
•	Bronchodilator Reversibility Test: Checks improvement in airflow after using bronchodilators.
•	Peak Flow Meter: Monitors daily lung function at home.
•	Methacholine Challenge Test: Confirms asthma diagnosis by provoking airway constriction.
•	Allergy Tests: Skin prick or blood tests to identify allergens that may trigger asthma.
•	Exhaled Nitric Oxide Test (FeNO): Measures airway inflammation.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from xgboost import XGBClassifier
from imblearn.over_sampling import SMOTE

# Load dataset
dataCBC = pd.read_csv('/content/cbc_dataset.csv')

dataCBC = dataCBC.dropna()  # Remove missing values

# Encode target variable
label_encoder = LabelEncoder()
dataCBC['Disease'] = label_encoder.fit_transform(dataCBC['Disease'])

# Feature Engineering
dataCBC['Hb_HCT_Ratio'] = dataCBC['Hemoglobin (Hb)'] / dataCBC['Hematocrit (%)']
dataCBC['RBC_WBC_Ratio'] = dataCBC['RBC Count (millions/\u00b5L)'] / dataCBC['WBC Count (thousands/\u00b5L)']

# Feature-target split
X = dataCBC.drop('Disease', axis=1)
y = dataCBC['Disease']

# Normalize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Stratified split to handle class imbalance
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, stratify=y, random_state=42)

# Apply SMOTE to balance classes
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

# Apply XGBoost Classifier
model = XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
model.fit(X_train_resampled, y_train_resampled)

y_pred = model.predict(X_test)

# Evaluation
accuracy = accuracy_score(y_test, y_pred)
print(f"XGBoost Accuracy: {accuracy:.2f}")
print("Classification Report:")
print(classification_report(y_test, y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

# Load dataset
dataCBC = pd.read_csv('/content/cbc_dataset.csv')

# Handle missing values (if any)
dataCBC = dataCBC.dropna()

# Encode the target variable
label_encoder = LabelEncoder()
dataCBC['Disease'] = label_encoder.fit_transform(dataCBC['Disease'])

# Split features and target
X = dataCBC.drop('Disease', axis=1)
y = dataCBC['Disease']

# Normalize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Initialize algorithms
models = {
    'Logistic Regression': LogisticRegression(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'Gradient Boosting': GradientBoostingClassifier(),
    'SVM': SVC(),
    'K-Nearest Neighbors': KNeighborsClassifier()
}

# Train and evaluate each model
results = {}
for name, model in models.items():
    print(f"\nTraining {name}...")
    model.fit(X_train, y_train)

    # Predictions
    y_pred = model.predict(X_test)

    # Evaluate
    accuracy = accuracy_score(y_test, y_pred)
    results[name] = accuracy

    print(f"Accuracy: {accuracy:.2f}")
    print("Classification Report:")
    print(classification_report(y_test, y_pred))
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))

# Display results
print("\nSummary of Model Performance:")
for name, accuracy in results.items():
    print(f"{name}: {accuracy:.2f}")
```

```python
from pdf2image import convert_from_path
import pytesseract
import re

# Step 1: Upload the PDF or Image File
from google.colab import files
print("Please upload the CBC report (PDF or image):")
uploaded = files.upload()

# Get the file path of the uploaded file
uploaded_file_path = list(uploaded.keys())[0]

# Step 2: Handle PDF to Image Conversion (if the uploaded file is a PDF)
if uploaded_file_path.endswith('.pdf'):
    images = convert_from_path(uploaded_file_path)
else:
    # If it's an image, load it directly
    from PIL import Image
    images = [Image.open(uploaded_file_path)]

# Step 3: Extract text from the images using OCR
def preprocess_text(text):
    # Fix common OCR errors and clean up text
    text = text.replace("Hemogtobin", "Hemoglobin").replace("R8C", "RBC").replace("W8C", "WBC")
    text = text.replace("Platelet Counf", "Platelet Count").replace("RDVV", "RDW").replace("RDV", "RDW")
    text = text.replace("g/dL", "").replace("x10^", "").replace("fL", "").replace("pg", "").replace("%", "")
    text = re.sub(r"[^\w\s\.\-:]", "", text)  # Remove special characters
    return text.strip()

extracted_text = ""
custom_config = r'--oem 3 --psm 6'  # Improve OCR performance
for img in images:
    extracted_text += pytesseract.image_to_string(img, config=custom_config)

# Preprocess the text for better matching
cleaned_text = preprocess_text(extracted_text)

# Debugging: Print the extracted text
####print("\nExtracted Text from Uploaded File:\n", cleaned_text)

# Step 4: Define a function to extract detailed CBC parameters
def extract_cbc_data(text):
    # Define regex patterns for CBC parameters
    parameters = {
        "Hemoglobin": r"(?:Hemoglobin|Hb)\s*[:\-]?\s*([\d\.]+)",
        "RBC Count": r"(?:RBC\s*(?:Count)?)\s*[:\-]?\s*([\d\.]+)",
        "TOTAL WBC Count": r"(?:TOTAL WBC\s*(?:Count)?)\s*[:\-]?\s*([\d\.]+)",
        "Platelet Count": r"(?:Platelet\s*(?:Count)?)\s*[:\-]?\s*([\d\.]+)",
        "Hematocrit": r"(?:Hematocrit|Hct)\s*[:\-]?\s*([\d\.]+)",
        "MCV": r"(?:MCV)\s*[:\-]?\s*([\d\.]+)",
        "MCH": r"(?:MCH)\s*[:\-]?\s*([\d\.]+)",
        "MCHC": r"(?:MCHC)\s*[:\-]?\s*([\d\.]+)",
        "RDW": r"(?:RDW-CV)\s*[:\-]?\s*([\d\.]+)"  # Adjusted RDW regex
    }

    results = {}
    for param, pattern in parameters.items():
        match = re.search(pattern, text, re.IGNORECASE)
        if match:
            try:
                results[param] = float(match.group(1))
            except ValueError:
                results[param] = match.group(1)
        else:
            print(f"DEBUG: '{param}' not found in text.")
            results[param] = None
    return results

# Step 5: Extract CBC data from the text
cbc_data = extract_cbc_data(cleaned_text)

# Debugging: Print extracted CBC parameters
print("\nExtracted Detailed CBC Data:")
####for key, value in cbc_data.items():
    ####print(f"{key}: {value}")

# Step 6: Store specific values in variables for further use
hemoglobin = cbc_data.get("Hemoglobin", None)
rbc_count = cbc_data.get("RBC Count", None)
wbc_count = cbc_data.get("TOTAL WBC Count", None)
platelet_count = cbc_data.get("Platelet Count", None)
hematocrit = cbc_data.get("Hematocrit", None)
mcv = cbc_data.get("MCV", None)
mch = cbc_data.get("MCH", None)
mchc = cbc_data.get("MCHC", None)
rdw = cbc_data.get("RDW", None)


wbc_count = wbc_count/1000
platelet_count = platelet_count/1000

print("\nKey CBC Variables:")
print(f"Hemoglobin: {hemoglobin}")
print(f"RBC Count: {rbc_count}")
print(f"WBC Count: {wbc_count}")
print(f"Platelet Count: {platelet_count}")
print(f"Hematocrit: {hematocrit}")
print(f"MCV: {mcv}")
print(f"MCH: {mch}")
print(f"MCHC: {mchc}")
print(f"RDW: {rdw}")
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier



# Normalize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Initialize algorithms
models = {
    'Logistic Regression': LogisticRegression(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'Gradient Boosting': GradientBoostingClassifier(),
    'SVM': SVC(),
    'K-Nearest Neighbors': KNeighborsClassifier()
}

# Train and evaluate each model
results = {}
for name, model in models.items():
    print(f"\nTraining {name}...")
    model.fit(X_train, y_train)

    # Predictions
    y_pred = model.predict(X_test)

    # Evaluate
    accuracy = accuracy_score(y_test, y_pred)*100
    results[name] = accuracy

    print(f"Accuracy: {accuracy:.2f}")
    print("Classification Report:")
    print(classification_report(y_test, y_pred))
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))

# Display results
print("\nSummary of Model Performance:")
for name, accuracy in results.items():
    print(f"{name}: {accuracy:.2f}")
```
