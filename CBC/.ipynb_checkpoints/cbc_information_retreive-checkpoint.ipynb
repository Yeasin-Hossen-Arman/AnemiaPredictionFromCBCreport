```python
from PIL import Image
import pytesseract
import re

# Load the image
image_path = '/content/1730716062.webp'
image = Image.open(image_path)

# Perform OCR on the image
ocr_result = pytesseract.image_to_string(image)

# Print OCR result (for debugging purposes)
print("OCR Result:\n", ocr_result)

# Define a function to extract specific parameters
def extract_cbc_parameters(text):
    # Parameters to extract
    parameters = ["WBC", "RBC", "Hemoglobin", "Hematocrit", "MCV", "MCH", "MCHC", "RDW", "Platelets"]

    # Dictionary to store extracted values
    results = {}

    for param in parameters:
        # Search for the parameter in the text
        match = re.search(fr"{param}\s+([\d\.]+)", text)
        if match:
            results[param] = float(match.group(1))  # Convert result to a float

    return results

# Extract the CBC parameters
cbc_results = extract_cbc_parameters(ocr_result)

# Print the extracted parameters
print("Extracted CBC Parameters:")
for key, value in cbc_results.items():
    print(f"{key}: {value}")

# Example usage in further processing
wbc = cbc_results.get("WBC", None)
rbc = cbc_results.get("RBC", None)
hemoglobin = cbc_results.get("Hemoglobin", None)

print("\nSpecific Values:")
print(f"WBC: {wbc}")
print(f"RBC: {rbc}")
print(f"Hemoglobin: {hemoglobin}")
```

```python
# Define normal reference ranges for CBC parameters
reference_ranges = {
    "Hemoglobin": (12.0, 16.0),  # Example range for females (g/dL)
    "RBC Count": (4.2, 5.9),     # (million/µL)
    "TOTAL WBC Count": (4.0, 11.0),  # (thousand/µL)
    "Platelet Count": (150.0, 450.0), # (thousand/µL)
    "Hematocrit": (37.0, 47.0),   # (%)
    "MCV": (80.0, 100.0),         # (fL)
    "MCH": (27.0, 33.0),          # (pg)
    "MCHC": (31.0, 36.0),         # (g/dL)
    "RDW": (11.5, 14.5)           # (%)
}

# Compare extracted CBC values against reference ranges
def compare_to_reference(cbc_data, reference_ranges):
    results = {}
    for param, (low, high) in reference_ranges.items():
        value = cbc_data.get(param)
        if value is None:
            results[param] = "Value not found"
        elif low <= value <= high:
            results[param] = f"Within Range ({low} - {high})"
        else:
            results[param] = f"Out of Range ({low} - {high})"
    return results

# Compare the extracted CBC data to reference ranges
comparison_results = compare_to_reference(cbc_data, reference_ranges)

# Print comparison results
print("\nComparison of Extracted CBC Data to Reference Ranges:")
for param, status in comparison_results.items():
    print(f"{param}: {status}")
```

```python
def extract_cbc_data_line_by_line(text):
    parameters = [
        "Hemoglobin", "RBC Count", "WBC Count", "Platelet Count",
        "Hematocrit", "MCV", "MCH", "MCHC", "RDW"
    ]

    results = {}
    lines = text.splitlines()

    for param in parameters:
        for line in lines:
            if param in line:
                # Extract the numeric value from the line
                match = re.search(r"([\d\.]+)", line)
                if match:
                    try:
                        results[param] = float(match.group(1))
                    except ValueError:
                        results[param] = match.group(1)
                break
        else:
            results[param] = None  # Mark as None if not found

    return results

cbc_data = extract_cbc_data_line_by_line(cleaned_text)
```

```python
from pdf2image import convert_from_path
import pytesseract
import re
from google.colab import files

# Step 1: Upload the PDF or Image File
print("Please upload the CBC report (PDF or image):")
uploaded = files.upload()

# Get the uploaded file path
uploaded_file_path = list(uploaded.keys())[0]

# Step 2: Handle PDF to Image Conversion (if the uploaded file is a PDF)
if uploaded_file_path.endswith('.pdf'):
    images = convert_from_path(uploaded_file_path)
else:
    # If it's an image, load it directly
    from PIL import Image
    images = [Image.open(uploaded_file_path)]

# Step 3: Extract text from the images using OCR
extracted_text = ""
for img in images:
    extracted_text += pytesseract.image_to_string(img)

# Debugging: Print the raw extracted text
print("\nExtracted Text from Uploaded File (Raw):\n", extracted_text)

# Step 4: Preprocess the extracted text to fix common OCR issues
def preprocess_text(text):
    # Replace common OCR errors
    cleaned_text = text.replace("R8C", "RBC").replace("W6C", "WBC")
    cleaned_text = cleaned_text.replace("PlateletCounf", "Platelet Count").replace("Hemogtobin", "Hemoglobin")
    return cleaned_text

cleaned_text = preprocess_text(extracted_text)

# Debugging: Print the cleaned text
print("\nCleaned Extracted Text:\n", cleaned_text)

# Step 5: Extract detailed CBC parameters using regex
def extract_cbc_data(text):
    # Define regex patterns for each parameter
    parameters = {
        "Hemoglobin": r"Hemoglobin\s*[:\-]?\s*([\d\.]+)",
        "RBC Count": r"RBC\s*(?:Count)?\s*[:\-]?\s*([\d\.]+)",
        "WBC Count": r"WBC\s*(?:Count)?\s*[:\-]?\s*([\d\.]+)",
        "Platelet Count": r"Platelet\s*(?:Count)?\s*[:\-]?\s*([\d\.]+)",
        "Hematocrit": r"Hematocrit\s*[:\-]?\s*([\d\.]+)",
        "MCV": r"MCV\s*[:\-]?\s*([\d\.]+)",
        "MCH": r"MCH\s*[:\-]?\s*([\d\.]+)",
        "MCHC": r"MCHC\s*[:\-]?\s*([\d\.]+)",
        "RDW": r"RDW\s*[:\-]?\s*([\d\.]+)"
    }

    # Extract parameter values
    results = {}
    for param, pattern in parameters.items():
        match = re.search(pattern, text)
        if match:
            try:
                results[param] = float(match.group(1))  # Convert to float
            except ValueError:
                results[param] = match.group(1)  # Keep as string if conversion fails
        else:
            results[param] = None  # Mark as None if not found

    return results

# Extract CBC data from the cleaned text
cbc_data = extract_cbc_data(cleaned_text)

# Step 6: Print extracted CBC parameters
print("\nExtracted Detailed CBC Data:")
for key, value in cbc_data.items():
    print(f"{key}: {value}")

# Step 7: Store specific values in variables for further use
hemoglobin = cbc_data.get("Hemoglobin", None)
rbc_count = cbc_data.get("RBC Count", None)
wbc_count = cbc_data.get("WBC Count", None)
platelet_count = cbc_data.get("Platelet Count", None)
hematocrit = cbc_data.get("Hematocrit", None)
mcv = cbc_data.get("MCV", None)
mch = cbc_data.get("MCH", None)
mchc = cbc_data.get("MCHC", None)
rdw = cbc_data.get("RDW", None)

# Step 8: Print the extracted key CBC variables
print("\nKey CBC Variables:")
print(f"Hemoglobin: {hemoglobin}")
print(f"RBC Count: {rbc_count}")
print(f"WBC Count: {wbc_count}")
print(f"Platelet Count: {platelet_count}")
print(f"Hematocrit: {hematocrit}")
print(f"MCV: {mcv}")
print(f"MCH: {mch}")
print(f"MCHC: {mchc}")
print(f"RDW: {rdw}")

```
